# TreebankWordTokenizer
Этот токенайзер разбивает строки на слова, так же обращаясь к word_tokenizer.
Особенности данного токенайзера:
> Разделяет стандартные сокращения, например don't -> do n't and they'll -> they 'll

> Рассматривает большинство знаков препинания как отдельные токены

> Отделяет запятые и одинарные кавычки, если за ними следует пробел

> Отделяет точки в конце предложений

```python
from nltk.tokenize import TreebankWordTokenizer
TreebankWordTokenizer().tokenize(str(sent_15))
```

# TweetTokenizer
этот токенизатор создан специально для твитов, он не удаляет хэш-тэги и знаки пунктуации

```python
from nltk.tokenize import TweetTokenizer
tknzr = TweetTokenizer()
tknzr.tokenize(str(sent_15))
```

# MWETokenizer
Использует словарь, для того чтобы объединять слова в один токен

```python
from nltk.tokenize import MWETokenizer
tokenizer = MWETokenizer([('a', 'little'), ('at', 'the'), ('a', 'lot')])
tokenizer.tokenize(words)
```

# WordPunctTokenizer
Разделяет текст на последовательности алфавитных и неалфавитных символов с помощью регулярного выражения \w+|[^\w\s]+.

```python
from nltk.tokenize import WordPunctTokenizer
WordPunctTokenizer().tokenize(str(sent_15))
```

### Существует еще множество токенизаторов, но я не стала их использовать, так как почти все они делают то же самое, что и .split()


*   nltk.tokenize.SpaceTokenizer — разделяет токены по пробелам аналогично s.split(' ').

*  nltk.tokenize.TabTokenizer — разделяет токены по символам табуляции аналогично s.split('\t').

*  nltk.tokenize.LineTokenizer — разделяет токены по символам переноса строки аналогично s.split('\n').


*  nltk.tokenize.WhitespaceTokenizer — разделяет токены по символам пробела, табуляции или переноса строки аналогично s.split().


*  nltk.tokenize.BlanklineTokenizer — разделяет токены используя в качестве разделителя любые последовательности пробелов и табуляций.

*   nltk.tokenize.RegexpTokenizer — разделяет текст с помощью регулярного выражения, которое необходимо самостоятельно задать.


*   nltk.tokenize.SExprTokenizer — разделяет текст, состоящий из S-выражений.













# Написать собственный токенизатор и сравнить его производительность по результату работы и времени выполнения токенизации применительно к тексту из Задания 1.

Мой токенизатор разбивает строки на слова, удаляет все знаки препинания и тп из строк, переводит слова в нижний регистр и выводит результат

```python
def my_own_super_tokenizer2000(text):
  word = text.split()
  for word in word:
    word = str(word).lower()  #перевод в нижний регистр
    word = re.sub(r'[^a-z0-9]', ' ', word) #выборока нужных символов
    print(word)
```
